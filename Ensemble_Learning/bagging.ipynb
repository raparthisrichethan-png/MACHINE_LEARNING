### Ensemble Techniques

That combine multiple models("usually weak learners") to produce a single and more accurate model than any other individual model 

Two types

a) Bagging

b) Boosting

### Bagging (Bootstrap aggregation)

It is an ensemble method where multiple models train independently on different random samples of dataset and their prediction are combined to produce to final result (multiple models work independetly)

### why bagging?

A single decision tree
    
    1. High variance

    2. Easily overfits

    3. Changes a lot if data changes silently

bagging reduces variance and makes prediction stable

### How bagging works

### bootstrap sampling

1) From the original dataset create multiple new datasets

2) Each data set is created by random sampling with replacement

3) Each bootstrap dataset has same size as the original 

### some data points repeat,some data points are left out this is called out of bag samples

### Train multiple models

1. Train one model for bootstraping dataset

2. Usually it is a decision tree

3. Each tree is trained independently 

4. Make predictions 

5. For regression take the average for classification take the majority vote

### Final output 

1. Aggregated prediction is more stable and accurate than a single decision tree

### When should we use bagging 

1. When dataset is medium to large

2. Model is overfitting

3. Decision tree gives unstable result
